**Redes LSTM**\
En los últimos años, ha habido un éxito increíble al aplicar RNN a una variedad de problemas como lo son el reconocimiento de voz, modelado de lenguaje, la traducción o los subtítulos de imágenes y la lista continua. Pero algo esencial para estos logros ha sido el uso de las “LSTM”, las cuales son un tipo de red neuronal recurrente que funciona para muchas tareas, mucho mejor que la versión estándar de las redes recurrentes.

Las LSTM están diseñadas específicamente para evitar el problema de la dependencia a largo plazo ya que recordar información durante largos periodos de tiempo es prácticamente su comportamiento predeterminado.

Las LSTM tienen al igual que las redes recurrentes una estructura similar a una cadena, pero el módulo repetitivo tiene una estructura diferente.

**La idea central detrás de las LSTM**\
La clave de las LSTM es el estado de la celda, esta se puede describir como una cinta transportadora la cual se ejecuta directamente a lo largo de toda la cadena de la red, con solo algunas iteraciones lineales menores. Es muy fácil que la información fluya. Las LSTM tienen además la capacidad de eliminar o agregar información al estado de la celda mediante unas estructurad denominadas puertas.

Ademas las LSTM permiten a las RNN el reconocimiento de sus entradas durante un largo período de tiempo. Esto se debe a que LSTM contiene su información en la memoria, que puede considerarse similar a la memoria de un ordenador , en el sentido que una neurona de una LSTM puede leer, escribir y borrar información de su memoria.

El reconocimiento de texto es una de las partes más importantes de una RNN, ya que partir de esta característica y la creatividad, podemos obtener la generación y de algún modo, puede ser considerada un paso previo. 

El texto se le conoce al conjunto de frases y palabras coherentes y ordenadas que permiten ser interpretadas y transmiten las ideas de un autor. Su extensión es variable y corresponde a un todo comprensible que tiene una finalidad comunicativa en un contexto dado. 

Las redes neuronales LSTM fueron una solución propuesta por Hochreiter y Schmidhuber [6] en 1997 al problema que se presentaba con las redes neuronales recurrentes (RNN, por sus siglas en ingles) sobre su poca capacidad de memoria, solucionando matemáticamente y algoritmicamente el problema al impedir los "vanishing gradients" (desvanecimiento de gradientes que provocaba una incapacidad de recordar eventos a largo plazo) que se hacia evidente en el "anidamiento" de los "hidden state" (capa oculta de memoria de las RNN) cuando se requería que se procesaran cantidades extensas de datos. Para lograr esto aprovecharon un "anidamiento" de redes neuronales dispuestas en un algoritmo entrenable que resultaba en poder seleccionar que eventos recordar a largo plazo y cuales olvidar segun la relevancia que la misma red consideraba. Dicho algoritmo se fundamenta en una segunda etapa o forma de memoria llamada "cell state" la cual ayuda a memorizar a largo plazo. El funcionamiento para filtrar la información que ingresara al cell state consta de compuertas o "gates" (por su termino en ingles) ademas de considerar el cell state previo. Las gates constan de lo siguiente; una "forget gate" (elemento que se encarga de proponer que elementos serán eliminados), una "input gate" (elemento que se encargara de proponer que tan memorable es un dato), por ultimo tenemos la update gate (que se encargara de filtrar que datos serán la salida en la cell state y hidden state.

Dichas compuertas funcionan normalmente con la función "sigmoide" debido a que la escala que presenta de 0 a 1 indica que tan importante o "olvidable" es un dato, ademas de que la función sigmoide. Cabe destacar que en la etapa de ingreso a la input gate se evaluan datos del hidden state con una función tangente debido a que el rango negativo (-1,1) puede también indicar el impacto no solo nulo si no a "contrapeso" de un dato para futuras referencias.

````python
# Send encoded text to Dataset PyTorch class
import torch
from torch.utils.data import Dataset
seq_length = 180 # fixed sequence length
chunk_size = seq_length + 1
text_chunks = [text_encoded[i:i+chunk_size] 
               for i in range(len(text_encoded)-chunk_size)]

class TextDataset(Dataset):
  def __init__(self, text_chunks):
    self.text_chunks = text_chunks

  def __len__(self):
    return len(self.text_chunks)

  def __getitem__(self, idx):
    text_chunk = self.text_chunks[idx]
    return text_chunk[:-1].long(), text_chunk[1:].long()

seq_dataset = TextDataset(torch.tensor(text_chunks))
````

````python
for i, (seq, target) in enumerate(seq_dataset):
  print(' Input (x): ',
    repr(''.join(char_array[seq])))
  print('Target (y): ',
    repr(''.join(char_array[target])))
  print()
  if i == 1:
    break
````

````python
# Define DataLoaders
from torch.utils.data import DataLoader
batch_size = 64
torch.manual_seed(1)
seq_dl = DataLoader(seq_dataset, batch_size=batch_size,
                    shuffle=True, drop_last=True)
````

## 4.2. Building the model

### LSTM Model

````python
# LSTM model definition
import torch.nn as nn
class RNN(nn.Module):
  def __init__(self, vocab_size, embed_dim, rnn_hidden_size):
    super().__init__()
    self.embedding = nn.Embedding(vocab_size, embed_dim)
    self.rnn_hidden_size = rnn_hidden_size
    #self.rnn = nn.RNN(input_size, hidden_size, num_layers=2, batch_first=True)
    self.rnn = nn.LSTM(embed_dim, rnn_hidden_size,
                       batch_first=True)
    self.fc = nn.Linear(rnn_hidden_size, vocab_size)

  def forward(self, x, hidden, cell):
    out = self.embedding(x).unsqueeze(1)
    out, (hidden, cell) = self.rnn(out, (hidden, cell))
    out = self.fc(out).reshape(out.size(0), -1)
    return out, hidden, cell

  def init_hidden(self, batch_size):
    hidden = torch.zeros(1, batch_size, self.rnn_hidden_size)
    cell = torch.zeros(1,batch_size, self.rnn_hidden_size)
    return hidden, cell
````

````python
# Model instantiation
vocab_size = len(char_array)
embed_dim = 256
rnn_hidden_size = 512
torch.manual_seed(1)
model = RNN(vocab_size, embed_dim, rnn_hidden_size).cuda()
model
````

### Training

````python
# Training criterions
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
````

````python
# Training
from tqdm import tqdm

num_epochs = 10000 #10000 -- Sofi
torch.manual_seed(1)
for epoch in tqdm(range(num_epochs)):
  hidden, cell = model.init_hidden(batch_size)
  hidden = hidden.cuda()
  cell = cell.cuda()
  seq_batch, target_batch = next(iter(seq_dl))
  seq_batch = seq_batch.cuda()
  target_batch = target_batch.cuda() 
  optimizer.zero_grad()
  loss = 0
  for c in range(seq_length):
    pred, hidden, cell = model(seq_batch[:, c], hidden, cell)
    loss += loss_fn(pred, target_batch[:, c])
  loss.backward()
  optimizer.step()
  loss = loss.item()/seq_length
  if epoch % 500 == 0:
    print(f'Epoch {epoch} loss: {loss:.4f}')
````

### Evaluation

````python
from torch.distributions.categorical import Categorical
torch.manual_seed(1)
logits = torch.tensor([[1.0, 1.0, 1.0]])
print('Probabilities:',
      nn.functional.softmax(logits, dim=1).numpy()[0])
m = Categorical(logits=logits)
samples = m.sample((10,))
print(samples.numpy())
````

## 4.3. Performing next-character prediction

````python
def sample(model, starting_str,len_generated_text=500,scale_factor=1.0):

  encoded_input = torch.tensor([char2int[s] for s in starting_str]).cuda()
  encoded_input = torch.reshape(encoded_input, (1, -1)).cuda()
  generated_str = starting_str
  model
  model.eval()

  hidden, cell = model.init_hidden(1)
  hidden = hidden.cuda()
  cell = cell.cuda()
  for c in range(len(starting_str)-1):
    _, hidden, cell = model(encoded_input[:, c].view(1), hidden, cell)
  hidden = hidden.cuda()
  cell = cell.cuda()
  last_char = encoded_input[:, -1]
  last_char = last_char.cuda()
  
  for i in range(len_generated_text):
    logits, hidden, cell = model(last_char.view(1), hidden, cell)
    logits = logits.cuda()
    logits = torch.squeeze(logits, 0)
    scaled_logits = logits * scale_factor
    m = Categorical(logits=scaled_logits)
    last_char = m.sample()
    generated_str += str(char_array[last_char])
  
  return generated_str
````
